```python
import io
import numpy as np
import matplotlib.pyplot as plt
```

```python
# We access the data set to gather sentiment values for a large number of words
f = open("Happiness.txt")
lines = f.readlines()
# We only want the lines corresponding to the words
lines = lines[4:]
# We create a list of the words and another containing their average happiness rank
sentimentWords = lines[:]
sentimentLevels = lines[:]
c = len(lines)
for i in range(c):
    s = lines[i].split()
    sentimentWords[i] = s[0]
    sentimentLevels[i] = float(s[2])
```

```python
# We now create the function
def sentiment(word):
    # We keep track of the total happiness rankings of the tokens and how many of them are applicable
    if word.lower() in sentimentWords:
        i = sentimentWords.index(word.lower())
        return sentimentLevels[i]
    else:
        return 0
```

```python
# We define a function that takes all the tokens and the window size.
# It returns a list of profiles for each moving average window.
def sentProfs(tokens, profSize, isOneSection):
    #Create a list for the average happiness level of each window
    sentProfiles = []
    # Keeps track of the sentiment values of the current window
    currentWindow = []
    # Keeps track of the total amount of happiness in the current window
    hapWinTot = 0
    # Takes the sentiment of the first word
    first = sentiment(tokens[0])
    # Keeps track of how many words in the current window have no sentiment value
    zeros = 0
    # Length of the list of words given
    tokenLen = len(tokens)
    #First set of values
    for i in range(profSize):
        n = sentiment(tokens[i])
        hapWinTot += n
        currentWindow.append(n)
        if(n == 0):
            zeros += 1
    # This boolean applies for the character interaction analysis
    if(isOneSection):
        return hapWinTot/(profSize - zeros)
    
    # place the first window in the list of profiles
    sentProfiles.append(hapWinTot/(profSize - zeros))

    # After the first window is analysed, we thereafter add a word to the window and remove the first word.
    # We update the values for finding the average sentiment of the window and then repeat the process until
    # we reach the end of the list of words.
    for i in range(profSize,tokenLen):
        if(currentWindow[0] == 0):
            zeros -= 1
        else:
            hapWinTot -= currentWindow[0]
        # Removes the first element in the window
        currentWindow.pop(0)
        
        n = sentiment(tokens[i])
        # Update the total happiness of the window
        hapWinTot += n
        currentWindow.append(n)
        if(n == 0):
            zeros += 1
            
        if(zeros == profSize):
            print "Whole window with no applicable words"
            continue
            
        sentProfiles.append(hapWinTot/(profSize - zeros))
        #if(hapWinTot/(profSize - zeros) < 2.84):
           # print currentWindow
        
  
    return sentProfiles
```

```python
# We make a list of words for each of the three books in the Lord of the Rings story line.
fotr_words = [word for line in open('LOTR_Fotr.txt', 'r') for word in line.split()]
tt_words = [word for line in open('LOTR_ttt.txt', 'r') for word in line.split()]
rotk_words = [word for line in open('LOTR_rotk.txt', 'r') for word in line.split()]
# We then compile the words from each book into one long list
All_words = fotr_words[:]
All_words.extend(tt_words)
All_words.extend(rotk_words)

# We also make a list of words based on the scripts for each movie along with a compiled version
fotr_script = [word for line in open('LordoftheRings1-FOTR-Script.txt', 'r') for word in line.split()]
tt_script = [word for line in open('LordoftheRings2-TTT-Script.txt', 'r') for word in line.split()]
rotk_script = [word for line in open('LordoftheRings3-ROTK-Script.txt', 'r') for word in line.split()]
All_scripts = fotr_script[:]
All_scripts.extend(tt_script)
All_scripts.extend(rotk_script)

print "Word Count in the Books: %s"%len(All_words)
print "Word Count in the Movie Scripts: %s"%len(All_scripts)

# This is how much more text there is in the books compared to the movie scripts
book_over_script_count_ratio = float(len(All_words))/float(len(All_scripts))
print "Book word count ratio over scripts: %f"%book_over_script_count_ratio
```

```python
# We want to have a graph of the movies and the books that's comparable. 
# The books have a lot more text than the scripts of the movies, but tell the same story.
# To see if the movies follow the same sentimental pattern as the books, we simply scale the 
# moving window of the book analysis up to compensate for the larger amounts of text. This
# makes the graph below for the books rouhgly as readable as that of the movies.
sp4000comp_words = sentProfs(All_words, int(4000.0 * book_over_script_count_ratio),bool(0))
sp4000_scripts = sentProfs(All_scripts, 4000,bool(0))
```

```python
# We plot the sentiment graphs of both the books and movies all the way through.
# The vertical dashed lines represent the three volumes.

plt.figure(num=1, figsize=(16, 6))
plt.plot(sp4000comp_words)
plt.ylabel('Happiness Rating')
plt.xlabel('Window Position in Text')
plt.axis([0, len(sp4000comp_words), 5.2, 5.45])
plt.title("Sentiment throughout the books")
plt.axvline(x=len(fotr_words), color='green',linestyle='--')
plt.axvline(x=len(fotr_words+tt_words), color='green',linestyle='--')
ax = plt.subplot()
ax.set_facecolor((0.176470588, 0.176470588,0.176470588))
#ax.spines['bottom'].set_color('white')
#ax.spines['left'].set_color('white')
#ax.yaxis.label.set_color('white')
#ax.xaxis.label.set_color('white')
#ax.tick_params(axis='x', colors='white')
#ax.tick_params(axis='y', colors='white')

plt.figure(num=2, figsize=(16, 6))
plt.plot(sp4000_scripts)
plt.ylabel('Happiness Rating')
plt.xlabel('Window Position in Text')
plt.axis([0, len(sp4000_scripts), 5.2, 5.45])
plt.title("Sentiment throughout the movies")
plt.axvline(x=len(fotr_script), color='green',linestyle='--')
plt.axvline(x=len(fotr_script+tt_script), color='green',linestyle='--')
ax = plt.subplot()
ax.set_facecolor((0.176470588, 0.176470588,0.176470588))
#ax.spines['bottom'].set_color('white')
#ax.spines['left'].set_color('white')
#ax.yaxis.label.set_color('white')
#ax.xaxis.label.set_color('white')
#ax.tick_params(axis='x', colors='white')
#ax.tick_params(axis='y', colors='white')

plt.show()
```

```python
# This function finds the sentiment of relationships between two characters throughout the books.
def charSent(char1, char2):
    indices1 = [i for i, x in enumerate(All_words) if x == char1]    
    indices2 = [i for i, x in enumerate(All_words) if x == char2]
    
    # For every instance of the first character's name, we see if the other character's 
    # name appears within a certain amount of words. 
    
    
    sentWindow = 500
    rad = int(sentWindow/2)
    
    # This keeps track of the sentiment for every interaction
    encounters = []
    # This keeps track of the position of every interaction
    encountersi = []
    
    for i in indices1:
        for j in range(i-rad,  i+rad):
            if j in indices2:
                # When two characters' names are close enough to eachother, we find the sentiment 
                # of all the words around the instance.
                if(i-rad < 0):
                    encounters.append(sentProfs(All_words[0 : 2*rad], sentWindow, bool(1)))
                elif (i+rad > len(All_words)):
                    encounters.append(sentProfs(All_words[:-sentWindow],sentWindow, bool(1)))
                else:
                    encounters.append(sentProfs(All_words[i-rad : i+rad], sentWindow, bool(1)))
                    
                encountersi.append(i)
                break
    #print encounters
    return [encountersi, encounters]
        
```

```python
import itertools
colors = ['skyblue','green','red','orange','yellow','purple','turquoise','gold']
# This function takes pairs of character names and plots their sentiment throughout the books.
def plotChar(charList, title):

    plt.figure(num=5, figsize=(16, 12))
    for idx, chars in enumerate(charList):
        charProf = charSent(chars[0],chars[1])
        plt.plot(charProf[0],charProf[1],color=colors[idx],marker='o', label=chars[0] + " - " + chars[1].decode('utf-8').encode('latin1'))
    ax = plt.subplot()
    ax.set_facecolor((0.176470588, 0.176470588,0.176470588))
    l = plt.legend()
    plt.ylabel('Happiness Rating')
    plt.xlabel('Word in Texts')
    plt.title(title, size=20)
    plt.axvline(x=len(fotr_words), color='green',linestyle='--')
    plt.axvline(x=len(fotr_words+tt_words), color='green',linestyle='--')
    #ax.spines['bottom'].set_color('white')
    #ax.spines['left'].set_color('white')
    #ax.yaxis.label.set_color('white')
    #ax.xaxis.label.set_color('white')
    #ax.tick_params(axis='x', colors='white')
    #ax.tick_params(axis='y', colors='white')
    plt.show()
```
